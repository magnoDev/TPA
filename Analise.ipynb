```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.gridspec as gridspec
import matplotlib.ticker as ticker


fig, ax = plt.subplots()
ax.grid(True)

ax.get_xaxis().get_major_formatter().set_useOffset(False)
ax.get_yaxis().get_major_formatter().set_useOffset(False)
head = ["algorithm", "size", "min", "max", "avg" ]
df = pd.read_csv('./src/output/records.dat')

df = df.sort_values(by='size')
insertion= df[df.algorithm=='insertion'].query('avg >= 0')
quick = df[df.algorithm=='quick'].query('avg >= 0').drop_duplicates('size')
merge = df[df.algorithm=='merge'].query('avg >= 0').drop_duplicates('size')
heap = df[df.algorithm=='heap'].query('avg >= 0').drop_duplicates('size')
selection = df[df.algorithm=='selection'].query('avg >= 0')
#plt.figure(figsize=(20,10))
plt.plot('size' , 'avg', data=insertion,label="insertion" )
plt.plot( 'size',  'avg', data=selection,label="selection")
plt.plot( 'size',  'avg', data=merge,label="merge")
plt.plot( 'size',  'avg', data=heap,label="heap")
plt.plot('size',  'avg', data=quick,label="quick")
plt.rcParams["figure.figsize"] = [20,10]
plt.xlabel("Tamanho do vetor")
plt.ylabel("Tempo (segundos)")
plt.title("Tempo de ordenao", fontsize=16, fontweight='bold')
plt.plot(usevlines=True,normed=True, maxlags=20)
plt.legend()
plt.savefig('./imgs/all.png',transparent=False)
```
